```python
#筛选，使得swot数据w-h正序超过一半

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

DRAW_FIGURE = False

df_s1all = pd.read_csv('swot_s1.csv')
stationids = df_s1all['stationid'].unique()

df_res = []
for s in stationids[:]:
    df_s1 = df_s1all[df_s1all['stationid']==s]
    df_s2 = df_s1all[df_s1all['stationid']==s]

    while True:
        n = len(df_s2)
        w = df_s2['width'].values
        h = df_s2['wse'].values
        w_diff = np.repeat(w,n).reshape(n,-1) - np.tile(w,(n,1))
        h_diff = np.repeat(h,n).reshape(n,-1) - np.tile(h,(n,1))
        wh = w_diff * h_diff
        inverse = np.count_nonzero(wh<0, axis=1)
        idx_max = np.argmax(inverse)
        if inverse[idx_max] / n < 0.5: break
        df_s2 = df_s2.drop(df_s2.index[idx_max])
    df_res.append(df_s2)


    if not DRAW_FIGURE: continue
    fig, ax = plt.subplots(1,2,figsize=(12,6))
    ax1, ax2 = ax
    ax1.scatter(df_s1['width'], df_s1['wse'], s=8, c='#BBBBBB', zorder=1)
    ax1.scatter(df_s2['width'], df_s2['wse'], s=8, c='red', zorder=1)
    ax1.errorbar(df_s1['width'], df_s1['wse'], xerr=df_s1['width_u'], yerr=df_s1['wse_u'],
                 fmt='none', lw=1, ecolor='#BBBBBB', zorder=0)
    ax2.scatter(df_s1['width'], df_s1['wse'], s=8, c='#BBBBBB', zorder=1)
    ax2.scatter(df_s2['width'], df_s2['wse'], s=8, c='red', zorder=1)
    ax2.errorbar(df_s1['width'], df_s1['wse'], xerr=df_s1['width_u'], yerr=df_s1['wse_u'],
                 fmt='none', lw=1, ecolor='#BBBBBB', zorder=0)
    ax1.set_xlabel('width (m)')
    ax1.set_ylabel('WSE (m)')
    ax2.set_xlabel('width (m)')
    ax2.set_ylabel('WSE (m)')
    xmin, xmax = df_s2['width'].min(), df_s2['width'].max()
    ymin, ymax = df_s2['wse'].min(), df_s2['wse'].max()
    ax2.set_xlim(xmin*1.1-xmax*0.1, xmax*1.1-xmin*0.1)
    ax2.set_ylim(ymin*1.1-ymax*0.1, ymax*1.1-ymin*0.1)
    plt.suptitle('%s retain: %d/%d'%(s,len(df_s2),len(df_s1)))
    fig.tight_layout()
    #plt.show()
    plt.savefig('./1_qc/2_order/%s.png'%s, dpi=200)

df_res = pd.concat(df_res)
df_res.to_csv('swot_s2.csv', index=False)
```
