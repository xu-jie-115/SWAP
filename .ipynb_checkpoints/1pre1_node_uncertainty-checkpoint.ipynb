```python
# 
# wse_u<0.4m wse_u_r<=0.1


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

DRAW_FIGURE = False

df_all = pd.read_csv('node_clip_0403.csv')
df_comid = pd.read_csv('gages3000_GRFR_q50_slp.csv')
df_all = df_all.replace(-999999999999, np.nan)
df_all = df_all.dropna()
df_comid = df_comid[['stationid','COMID']]
df_all = df_all.merge(df_comid, on='stationid', how='left')
print(df_all.head())

df_all['node_data_num'] = df_all.groupby('node_id')['lat'].transform('count')
df_all = df_all[df_all['node_data_num']>=10]  # retain nodes with >= 10 swot data
stationids = df_all['stationid'].unique()

# select rank_corr max node
df_node = df_all[['node_id','stationid']].drop_duplicates(subset='node_id')
df_node = df_node.set_index('node_id')
for node in df_node.index:
    df_sel = df_all[df_all['node_id']==node]
    df_node.loc[node,'rank_corr'] = spearmanr(df_sel['width'], df_sel['wse'])[0]
df_node = df_node.reset_index()
max_idx = df_node.groupby('stationid')['rank_corr'].idxmax()
df_node_rmax = df_node.loc[max_idx]
df_all = df_all[df_all['node_id'].isin(df_node_rmax['node_id'])]

df_res = []
for s in stationids[:]:
    df = df_all[df_all['stationid']==s]
    df['width_u_r'] = df['width_u'] / df['width']
    df_sel = df[(df['wse_u']<=0.4) & (df['width_u_r']<=0.1)]
    df_res.append(df_sel)


    if not DRAW_FIGURE: continue
    fig, ax = plt.subplots(1,2,figsize=(12,6))
    ax1, ax2 = ax
    ax1.scatter(df['width'], df['wse'], s=8, c='#999999', zorder=1)
    ax1.scatter(df_sel['width'], df_sel['wse'], s=8, c='red', zorder=1)
    ax1.errorbar(df['width'], df['wse'], xerr=df['width_u'], yerr=df['wse_u'],
                 fmt='none', lw=1, ecolor='#999999', zorder=0)
    ax2.scatter(df_sel['width'], df_sel['wse'], s=8, c='red', zorder=1)
    ax2.errorbar(df_sel['width'], df_sel['wse'], xerr=df_sel['width_u'], yerr=df_sel['wse_u'],
                 fmt='none', lw=1, ecolor='#999999', zorder=0)
    ax1.set_xlabel('width (m)')
    ax1.set_ylabel('WSE (m)')
    ax2.set_xlabel('width (m)')
    ax2.set_ylabel('WSE (m)')
    plt.suptitle('%s retain: %d/%d'%(s,len(df_sel),len(df)))
    fig.tight_layout()
    #plt.show()
    plt.savefig('1_qc/1_uncertainty/1pre1_dule04_wurle01_%s.png'%s, dpi=200)

df_res = pd.concat(df_res)
df_res = df_res.drop_duplicates(subset=['stationid','time'], keep='first')
df_res[['COMID','stationid','node_id','wse','wse_u','width','width_u','time']].to_csv('swot_s1.csv', index=False)
```
