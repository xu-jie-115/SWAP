```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 16
plt.rcParams['axes.titlesize'] = 24
plt.rcParams['axes.labelsize'] = 24
plt.rcParams['axes.titlepad'] = 10
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['xtick.major.width'] = 1.6
plt.rcParams['ytick.major.width'] = 1.6


def load_and_preprocess_data(file_paths):
    """
    
    """
    bam_df = pd.read_csv(file_paths['bam'])
    geobam_df = pd.read_csv(file_paths['geobam'])
    w83_df = pd.read_csv(file_paths['w83'])
    swap_df = pd.read_csv(file_paths['swap'])
    slp_df = pd.read_csv(file_paths['slope'])

    # 
    for df in [bam_df, geobam_df, w83_df]:
        df.rename(columns={'PB': 'pBIAS'}, inplace=True)

    # 
    for df in [bam_df, geobam_df]:
        df['pBIAS'] /= 100.0

    # 
    for df in [bam_df, geobam_df, w83_df, swap_df]:
        df.dropna(inplace=True)

    return bam_df, geobam_df, w83_df, swap_df, slp_df


def filter_data_by_stationids(bam_df, geobam_df, w83_df, swap_df, slp_df):
    """
     stationid  stationid 
    """
    common_ids = set(bam_df['stationid']) & set(geobam_df['stationid']) & \
                 set(w83_df['stationid']) & set(swap_df['stationid']) & set(slp_df['stationid'])

    for df in [bam_df, geobam_df, w83_df, swap_df]:
        df = df[df['stationid'].isin(common_ids)].sort_values(by='stationid').reset_index(drop=True)
    
    return bam_df, geobam_df, w83_df, swap_df


def plot_histogram(data, q25, q75):
    """
     25%  75% 
    """
    plt.figure(figsize=(10, 6))
    n, bins, patches = plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)

    plt.axvline(q25, color='red', linestyle='--', linewidth=2, label=f'25% Quantile = {q25:.4f}')
    plt.axvline(q75, color='green', linestyle='--', linewidth=2, label=f'75% Quantile = {q75:.4f}')
    
    plt.text(q25, max(n)*0.9, f'{q25:.4f}', color='red', ha='right', va='center', fontsize=10)
    plt.text(q75, max(n)*0.9, f'{q75:.4f}', color='green', ha='left', va='center', fontsize=10)

    plt.title("Histogram with 25% and 75% Quantiles")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.legend()
    plt.tight_layout()
    plt.show()


def plot_cdf(data, ax, label, x_range=None, color=None):
    """
     CDF 
    """
    sorted_data = sorted(data)
    cdf = [i / len(sorted_data) for i in range(1, len(sorted_data) + 1)]
    ax.plot(sorted_data, cdf, label=label, color=color)

    if x_range:
        ax.set_xlim(x_range)

    ax.set_ylabel('CDF')


def plot_comparative_cdf(bam_df, geobam_df, w83_df, swap_df, columns, x_axis_ranges):
    """
     CDF 
    """
    fig, axes = plt.subplots(2, 3, figsize=(18, 10))
    colors = {
        'BAM': 'blue',
        'Geobam': 'green',
        'W83': 'red',
        'SWAP': 'orange'
    }
    
    # 
    datasets = [
        (bam_df, 'BAM'),
        (geobam_df, 'Geobam'),
        (w83_df, 'W83'),
        (swap_df, 'SWAP')
    ]
    
    for i, col in enumerate(columns):
        ax = axes[i // 3, i % 3]

        #  dataset 
        for dataset, label in datasets:
            if col == 'pBIAS':
                plot_cdf(np.abs(dataset[col]), ax, label, x_range=x_axis_ranges[col], color=colors[label])
            elif col == 'RV':
                plot_cdf(np.abs(dataset[col] - 1), ax, label, x_range=x_axis_ranges[col], color=colors[label])
            else:
                plot_cdf(dataset[col], ax, label, x_range=x_axis_ranges[col], color=colors[label])

        ax.legend()
        ax.set_title(col)

    plt.tight_layout()
    plt.show()



def plot_heatmap(com_df):
    """
     heatmap
    """
    heatmap_data = com_df[['stationid', 'KGE', 'NSE', 'CC', 'NRMSE', '|pBIAS|', '|RV-1|']]
    heatmap_data = heatmap_data.set_index('stationid').reset_index()

    plt.figure(figsize=(10, 6))
    ax = sns.heatmap(heatmap_data.set_index('stationid'), cmap='bwr', annot=False, center=0, 
                     vmin=-1, vmax=1, cbar_kws={'label': 'Value', 'extend': 'both'}, yticklabels=False)
    plt.title('Heatmap of CC, nRMSE, KGE, NSE across stationid')
    plt.tight_layout()
    plt.show()


def calculate_positive_negative_proportions(df_metrics):
    """
    
    """
    metrics_positive = ['KGE', 'NSE', 'CC']
    metrics_negative = ['NRMSE', '|pBIAS|', '|RV-1|']

    proportion_positive = (df_metrics[metrics_positive] > 0).mean()
    proportion_negative = (df_metrics[metrics_negative] < 0).mean()

    print("Proportion of positive values for 'KGE', 'NSE', 'CC':")
    print(proportion_positive.round(2))
    print("\nProportion of negative values for 'NRMSE', '|pBIAS|', '|RV-1|':")
    print(proportion_negative.round(2))


def main():
    # 
    file_paths = {
        'bam': 'bamgeobam/statistics_all_bam.csv',
        'geobam': 'bamgeobam/statistics_all_geobam.csv',
        'w83': 'validation_W83_datematch.csv',
        # 'swap': 'validation_SWAP_datematch.csv',
        'swap' : 'validation_SWAP_datematch_modified_q50.csv',
        'slope': 'gages3000_GRFR_q50_slp.csv'
    }

    bam_df, geobam_df, w83_df, swap_df, slp_df = load_and_preprocess_data(file_paths)

    slp_data = slp_df['slope'].dropna()
    # slp25 = np.percentile(slp_data, 25)
    # slp75 = np.percentile(slp_data, 75)    
    # slp_df = slp_df[slp_df['slope']<slp75] 
    # # 
    # plot_histogram(slp_data, slp25, slp75)
    
    #  'stationid' 
    bam_df, geobam_df, w83_df, swap_df = filter_data_by_stationids(bam_df, geobam_df, w83_df, swap_df,slp_df)

    # 
    com_df = swap_df[['KGE', 'NSE', 'NRMSE', 'CC']] - w83_df[['KGE', 'NSE', 'NRMSE', 'CC']]
    com_df['|pBIAS|'] = abs(swap_df[['pBIAS']]) - abs(w83_df[['pBIAS']])
    com_df['|RV-1|'] = abs(swap_df[['RV']] - 1) - abs(w83_df[['RV']] - 1)
    com_df['stationid'] = w83_df['stationid']
    com_df = com_df.sort_values(by='KGE')

    #  CDF 
    columns = ['KGE', 'NSE', 'NRMSE', 'CC', 'pBIAS', 'RV']
    x_axis_ranges = {
        'KGE': [-1, 1],
        'NSE': [-1, 1],
        'NRMSE': [0, 1],
        'CC': [0, 1],
        'pBIAS': [0, 1],
        'RV': [0, 2]
    }
    plot_comparative_cdf(bam_df, geobam_df, w83_df, swap_df, columns, x_axis_ranges)

    #  heatmap
    plot_heatmap(com_df)

    # 
    calculate_positive_negative_proportions(com_df)


if __name__ == "__main__":
    main()

```

```python
import pandas as pd
import numpy as np
from scipy.stats import spearmanr 
from sklearn.metrics import r2_score, mean_squared_error

wq_df = pd.read_csv('3\q_kge_med.csv')
stations = wq_df['stationid'].unique()
res = []
for s in stations:
    station_data = wq_df[wq_df['stationid'] == s]
    corr, _ = spearmanr(station_data['qobs'], station_data['width'])
    res.append({'stationid': s, 'rank_correlation': corr})

#  DataFrame
corr_df = pd.DataFrame(res)

df_swot_all = pd.read_csv('swot_s3.csv')
df_fit_all = pd.read_csv('3/fit_proba.csv')
df_med_all = pd.read_csv('3/hypso_med.csv')
stations = sorted(df_med_all['stationid'].unique())


nrmse_df = pd.DataFrame({'stationid':stations}, index=stations)
for s in stations:
    df_swot = df_swot_all[df_swot_all['stationid']==s]
    
    df_fit = df_fit_all[df_fit_all['stationid']==s]
    df_med = df_med_all[df_med_all['stationid']==s]
    
    w_low, w_high = df_fit.iloc[0][['w_low','w_high']]
    d_bankfull = 0.27 * (w_high / 7.2)**0.6
    df_swot = df_swot[(df_swot['width']>w_low)&(df_swot['width']<w_high)]
    
    df_swot['wse_hypso'] = np.interp(df_swot['width'], df_med['width'], df_med['wse'])
    nrmse_df.loc[s,'r2'] = r2_score(df_swot['wse'], df_swot['wse_hypso'])
    nrmse_df.loc[s,'rmse'] = mean_squared_error(df_swot['wse'], df_swot['wse_hypso'])**0.5
    nrmse_df.loc[s,'fit_nrmse'] = nrmse_df.loc[s,'rmse'] / d_bankfull


df_qa = nrmse_df.merge(corr_df,on = 'stationid',how = 'inner')
df_qa = com_df.merge(df_qa,on='stationid',how='inner')
# df_qa = df_qa[df_qa['rank_correlation']>0.5]
# df_qa = df_qa[df_qa['fit_nrmse']<1]
df_qa = df_qa.dropna()
```
