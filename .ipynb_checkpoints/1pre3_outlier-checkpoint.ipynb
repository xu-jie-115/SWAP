```python
# 去掉高低点之外的width, 丢掉上下一个bankfull的h

import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')


DRAW_FIGURE = False

df_s1all = pd.read_csv('swot_s1.csv')
df_s2all = pd.read_csv('swot_s2.csv')
df_w = pd.read_csv('1_3_w50.csv', index_col='stationid')
stationids = df_s1all['stationid'].unique()

df_res = []
for s in stationids[:]:
    df_s1 = df_s1all[df_s1all['stationid']==s]
    df_s2 = df_s2all[df_s2all['stationid']==s]
    if s not in df_w.index: continue
    w_low, w_high = df_w.loc[s,['w_low','w_high']]
    d_bankfull = 0.27 * (w_high / 7.2)**0.6
    h50 = df_s2['wse'].median()
    std = df_s2['wse'].std()
    df_s3 = df_s2[(df_s2['wse']<=h50+d_bankfull) & (df_s2['wse']>=h50-d_bankfull)]
    df_res.append(df_s3)


    if not DRAW_FIGURE: continue
    xmax = df_s1['width'].max() * 1.01

    fig, ax = plt.subplots(1,2,figsize=(12,6))
    ax1, ax2 = ax
    ax1.scatter(df_s1['width'], df_s1['wse'], s=8, c='#BBBBBB', zorder=1)
    ax1.scatter(df_s3['width'], df_s3['wse'], s=8, c='red', zorder=1)
    ax1.errorbar(df_s1['width'], df_s1['wse'], xerr=df_s1['width_u'], yerr=df_s1['wse_u'],
                 fmt='none', lw=1, ecolor='#BBBBBB', zorder=0)
    ax1.plot([0,xmax], [h50,h50], lw=1, zorder=-1)
    ax1.fill_between([0,xmax], [h50-d_bankfull,h50-d_bankfull], [h50+d_bankfull,h50+d_bankfull], alpha=0.1, zorder=-2)
    ax1.plot([w_low,w_low], [h50-d_bankfull,h50+d_bankfull], c='black', lw=1, alpha=0.2, zorder=-1)
    ax1.plot([w_high,w_high], [h50-d_bankfull,h50+d_bankfull], c='black', lw=1, alpha=0.2, zorder=-1)
    ax2.scatter(df_s1['width'], df_s1['wse'], s=8, c='#BBBBBB', zorder=1)
    ax2.scatter(df_s2['width'], df_s2['wse'], s=8, c='red', zorder=1)
    ax2.errorbar(df_s1['width'], df_s1['wse'], xerr=df_s1['width_u'], yerr=df_s1['wse_u'],
                 fmt='none', lw=1, ecolor='#BBBBBB', zorder=0)
    ax2.plot([0,xmax], [h50,h50], lw=1, zorder=-1)
    ax2.fill_between([0,xmax], [h50-std,h50-std], [h50+std,h50+std], alpha=0.1, zorder=-2)
    ax2.fill_between([0,xmax], [h50-2*std,h50-2*std], [h50+2*std,h50+2*std], color='#1f77b4', alpha=0.1, zorder=-2)
    ax2.plot([w_low,w_low], [h50-2*std,h50+2*std], c='black', lw=1, alpha=0.2, zorder=-1)
    ax2.plot([w_high,w_high], [h50-2*std,h50+2*std], c='black', lw=1, alpha=0.2, zorder=-1)
    ax1.set_xlabel('width (m)')
    ax1.set_ylabel('WSE (m)')
    ax2.set_xlabel('width (m)')
    ax2.set_ylabel('WSE (m)')
    xmin, xmax = 0, df_s2['width'].max()
    ymin, ymax = df_s2['wse'].min(), df_s2['wse'].max()
    ax1.set_xlim(xmin*1.1-xmax*0.1, xmax*1.1-xmin*0.1)
    ax1.set_ylim(ymin*1.1-ymax*0.1, ymax*1.1-ymin*0.1)
    ax2.set_xlim(xmin*1.1-xmax*0.1, xmax*1.1-xmin*0.1)
    ax2.set_ylim(ymin*1.1-ymax*0.1, ymax*1.1-ymin*0.1)
    #plt.suptitle('%s retain: %d/%d'%(s,len(df_s3),len(df_s1)))
    fig.tight_layout()
    #plt.show()
    plt.savefig('1_qc/3_outlier/%s.png'%s, dpi=200)

df_res = pd.concat(df_res)
df_res.insert(0, 'COMID', df_res.pop('COMID'))
df_res.to_csv('swot_s3.csv', index=False)
```
