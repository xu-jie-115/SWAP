```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as patches
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr


def KGE(y_true, y_pred):  # improved 2012
    correlation = np.corrcoef(y_true, y_pred)[0, 1]
    alpha = (np.std(y_pred)/np.mean(y_pred)) / (np.std(y_true)/np.mean(y_true))
    beta = np.mean(y_pred) / np.mean(y_true)
    return 1 - np.sqrt((correlation - 1)**2 + (alpha - 1)**2 + (beta - 1)**2)

# 定义计算NSE的函数
def NSE(observed, simulated):
    return 1 - np.sum((observed - simulated) ** 2) / np.sum((observed - np.mean(observed)) ** 2)

# 定义计算rRMSE的函数
def nRMSE(observed, simulated):
    rmse = np.sqrt(mean_squared_error(observed, simulated))
    return rmse / np.mean(observed)

# 定义计算CC的函数
def CC(observed, simulated):
    return pearsonr(observed, simulated)[0]  # 使用Spearman相关系数作为CC


def swap_valid(swap_input_data):
    df = pd.read_csv(swap_input_data)
    
    stationids = df['stationid'].unique()
    valid = []
    
    for s in stationids:
        df_station = df[df['stationid']==s]
        observed = df_station['qobs']
        simulated = df_station['Q_est']
        cc = CC(observed, simulated)
        nrmse = nRMSE(observed, simulated)
        kge = KGE(observed, simulated)
        nse = NSE(observed, simulated)
        pbias = (simulated.mean() / observed.mean() - 1) 
        rv    = simulated.std() / simulated.mean()/ (observed.std() / observed.mean())
        valid.append({'stationid': s, 'KGE': kge, 'NSE': nse,'NRMSE': nrmse,'CC': cc, 'pBIAS': pbias, 'RV': rv})
    
    swap_df = pd.DataFrame(valid)
    return swap_df


def plot_cdf(data, ax, label, x_range=None, color=None):
    sorted_data = np.sort(data)
    cdf = np.arange(1, len(sorted_data)+1) / len(sorted_data)
    ax.plot(sorted_data, cdf, label=label, color=color)
    ax.grid(True)
    if x_range is not None:
        ax.set_xlim(x_range)


def compare_cdf(swap_input_csv, swap_output_csv, figure_name):
    # 读取数据
    bam_df = pd.read_csv('bamgeobam/statistics_all_bam.csv')
    geobam_df = pd.read_csv('bamgeobam/statistics_all_geobam.csv')
    w83_df = pd.read_csv('validation_W83.csv')
    swap_df = swap_valid(swap_input_csv)

    # 先统一列名
    for df in [bam_df, geobam_df, w83_df]:
        df.rename(columns={'PB': 'pBIAS'}, inplace=True)
    
    # 然后再做单位转换（如果有必要）
    for df in [bam_df, geobam_df]:
        df['pBIAS'] = df['pBIAS'] / 100.0
    
    # 最后移除缺失值
    for df in [bam_df, geobam_df, w83_df, swap_df]:
        df.dropna(inplace=True)

    
    # 统一列名
    columns = ['KGE', 'NSE', 'NRMSE', 'CC', 'pBIAS', 'RV']
    for df in [bam_df, geobam_df, w83_df, swap_df]:
        df = df[columns]

    # 筛选 stationid 交集
    common_ids = set(bam_df['stationid']) & set(geobam_df['stationid']) & set(w83_df['stationid']) & set(swap_df['stationid'])
    bam_df = bam_df[bam_df['stationid'].isin(common_ids)].reset_index(drop=True)
    geobam_df = geobam_df[geobam_df['stationid'].isin(common_ids)].reset_index(drop=True)
    w83_df = w83_df[w83_df['stationid'].isin(common_ids)].reset_index(drop=True)
    swap_df = swap_df[swap_df['stationid'].isin(common_ids)].reset_index(drop=True)

    # # 保存精度指标结果
    # swap_df.to_csv(swap_output_csv, index=False)

    # 设置绘图样式
    sns.set(style="whitegrid")
    fig, axes = plt.subplots(2, 3, figsize=(14, 8))
    colors = {
        'BAM': 'red',
        'Geobam': 'orange',
        'W83': 'green',
        'SWAP': 'blue'
    }

    x_axis_ranges = {
        'KGE': [-1, 1],
        'NSE': [-1, 1],
        'NRMSE': [0, 1],
        'CC': [0, 1],
        'pBIAS': [-1, 1],
        'RV': [0, 2]
    }

    for i, col in enumerate(columns):
        ax = axes[i//3, i%3]
        plot_cdf(bam_df[col], ax, 'BAM', x_range=x_axis_ranges[col], color=colors['BAM'])
        plot_cdf(geobam_df[col], ax, 'Geobam', x_range=x_axis_ranges[col], color=colors['Geobam'])
        plot_cdf(w83_df[col], ax, 'W83', x_range=x_axis_ranges[col], color=colors['W83'])
        plot_cdf(swap_df[col], ax, 'SWAP', x_range=x_axis_ranges[col], color=colors['SWAP'])
        ax.legend()
        ax.set_title(col)

    plt.tight_layout()
    # plt.savefig(figure_name, dpi=500)
    plt.show()

# ------------------ 批量调用 ------------------

if __name__ == "__main__":
    compare_cdf(
        swap_input_csv='3/q_kge_med_q50weighted.csv',
        swap_output_csv='validation_SWAP_q50weighted.csv',
        figure_name='comparing-CDF-q50weighted.png'
    )

    # compare_cdf(
    #     swap_input_csv='3/q_kge_med_modified.csv',
    #     swap_output_csv='validation_SWAP_modified.csv',
    #     figure_name='comparing-CDF-modified.png'
    # )

    # compare_cdf(
    #     swap_input_csv='3/q_kge_med_modified_q50.csv',
    #     swap_output_csv='validation_SWAP_modified_q50.csv',
    #     figure_name='comparing-CDF-modified_q50.png'
    # )
```
