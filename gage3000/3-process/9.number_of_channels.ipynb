```python
import pandas as pd

# 读取数据
df_attr = pd.read_csv('5.q_kge_med_modified_q50_iqr_node_1.5_noqa_VersionD.csv')
# 确保读取正确的列
df_attr = df_attr[['stationid', 'kge', 'nse', 'nrmse']].drop_duplicates(subset='stationid')
list_stationids = df_attr['stationid'].unique()

df_node = pd.read_csv('1.all_matched_points_VersionD.csv')
# 确保读取正确的列
df_node = df_node[['p_n_ch_max', 'p_n_ch_mod', 'stationid']].drop_duplicates(subset='stationid')

# 修正：过滤包含在list_stationids中的站点
df_node = df_node[df_node['stationid'].isin(list_stationids)]

# 筛选p_n_ch_max > 1的记录
max_df = df_node[df_node['p_n_ch_max'] > 1]
print(f"p_n_ch_max > 1 的记录数: {len(max_df)}")

# 合并KGE信息来计算KGE>0的占比
max_df_with_kge = max_df.merge(df_attr[['stationid', 'kge']], on='stationid', how='left')
kge_positive_count = (max_df_with_kge['kge'] > 0).sum()
kge_positive_ratio = kge_positive_count / len(max_df) if len(max_df) > 0 else 0
print(f"max_df中KGE>0的记录数: {kge_positive_count}")
print(f"max_df中KGE>0的占比: {kge_positive_ratio:.2%}")

# 筛选p_n_ch_mod > 1的记录
mod_df = df_node[df_node['p_n_ch_mod'] > 1]
print(f"\np_n_ch_mod > 1 的记录数: {len(mod_df)}")

# 合并KGE信息来计算KGE>0的占比
mod_df_with_kge = mod_df.merge(df_attr[['stationid', 'kge']], on='stationid', how='left')
kge_positive_count_mod = (mod_df_with_kge['kge'] > 0).sum()
kge_positive_ratio_mod = kge_positive_count_mod / len(mod_df) if len(mod_df) > 0 else 0
print(f"mod_df中KGE>0的记录数: {kge_positive_count_mod}")
print(f"mod_df中KGE>0的占比: {kge_positive_ratio_mod:.2%}")
```
