```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import spearmanr
from pathlib import Path

import warnings
warnings.filterwarnings('ignore')

DRAW_FIGURE = False

folder = Path('./1.swot-selection-fig')

if not folder.exists():
    folder.mkdir(parents=True)
    print(f"âœ… Created directory: {folder}")
else:
    print(f"ğŸ“ Directory already exists: {folder}")

target_stations = ['Australia_121001', 'Australia_208004', 'Australia_710003', 'Australia_809316', 'Brazil_10200000', 'Brazil_12540000', 'Brazil_12600001', 'Brazil_13170000', 'Brazil_13310000', 'Brazil_14325000', 'Brazil_14540000', 'Brazil_14750000', 'Brazil_14850000', 'Brazil_15120001', 'Brazil_15556010', 'Brazil_15795000', 'Brazil_15910000', 'Brazil_15930000', 'Brazil_16368000', 'Brazil_17090000', 'Brazil_18119000', 'Brazil_18200000', 'Brazil_18640000', 'Brazil_21600000', 'Brazil_24750000', 'Brazil_27110000', 'Brazil_34770000', 'Brazil_34790000', 'Brazil_34980000', 'Brazil_35170000', 'Brazil_36320000', 'Brazil_37470000', 'Brazil_37710150', 'Brazil_37730000', 'Brazil_37761000', 'Brazil_40150000', 'Brazil_43670000', 'Brazil_51350000', 'Brazil_52570000', 'Brazil_53170000', 'Brazil_53180000', 'Brazil_54002990', 'Brazil_55560000', 'Brazil_55630000', 'Brazil_56891900', 'Brazil_58128100', 'Brazil_58152010', 'Brazil_58158100', 'Brazil_58204000', 'Brazil_58630002', 'Brazil_58636000', 'Brazil_60030000', 'Brazil_60040000', 'Brazil_60510010', 'Brazil_61202000', 'Brazil_61425000', 'Brazil_61510000', 'Brazil_61787000', 'Brazil_61830000', 'Brazil_61884002', 'Brazil_63350100', 'Brazil_63900001', 'Brazil_64447000', 'Brazil_64771500', 'Brazil_65775901', 'Brazil_65815050', 'Brazil_65925000', 'Brazil_65992500', 'Brazil_66050000', 'Brazil_66231000', 'Brazil_66845000', 'Brazil_72430000', 'Brazil_72715000', 'Brazil_72810000', 'Brazil_72849000', 'Brazil_74210000', 'Brazil_77500000', 'Brazil_81679000', 'Brazil_86298000', 'Brazil_86410000', 'Brazil_86440000', 'Brazil_86447000', 'Brazil_87590000', 'Canada_01AD003', 'Canada_01AG003', 'Canada_01BO001', 'Canada_02GA016', 'Canada_02GB001', 'Canada_02GD015', 'Canada_03QC002', 'Canada_05AA024', 'Canada_05AD007', 'Canada_05BA001', 'Canada_05CC002', 'Canada_05TE002', 'Canada_06CC001', 'Canada_08LA001', 'Canada_08LC002', 'Canada_08NA006', 'Canada_10RA002', 'Canada_10UH001', 'Chile_05410002', 'Chile_05423003', 'Chile_06008005', 'Chile_06019003', 'Chile_07103001', 'Chile_07104002', 'Chile_07112001', 'Chile_07115001', 'Chile_07308002', 'Chile_07321002', 'Chile_07322001', 'Chile_07354001', 'Chile_07354002', 'Chile_07379002', 'Chile_08106002', 'Chile_08307002', 'Chile_08312000', 'Chile_08312001', 'Chile_08317004', 'Chile_08317005', 'Chile_08380002', 'Chile_08381003', 'Chile_09129002', 'Chile_09140001', 'Chile_09402001', 'EWA_9110141', 'EWA_9110287', 'EWA_9114186', 'EWA_9114213', 'EWA_9114217', 'EWA_9114248', 'EWA_9114378', 'EWA_9115148', 'EWA_9115360', 'EWA_9116051', 'EWA_9931006', 'GRDC_1112320', 'GRDC_1134220', 'GRDC_1309675', 'GRDC_1428400', 'GRDC_1531050', 'GRDC_1531100', 'GRDC_1591402', 'GRDC_1634500', 'GRDC_1813450', 'GRDC_1813460', 'GRDC_2181951', 'GRDC_2469055', 'GRDC_2587082', 'GRDC_2587083', 'GRDC_2587105', 'GRDC_2588302', 'GRDC_2588303', 'GRDC_2588322', 'GRDC_2588630', 'GRDC_2588653', 'GRDC_2588704', 'GRDC_2588708', 'GRDC_2588910', 'GRDC_2589240', 'GRDC_2589310', 'GRDC_2589320', 'GRDC_2589351', 'GRDC_2589360', 'GRDC_2589390', 'GRDC_2589501', 'GRDC_2589506', 'GRDC_2589703', 'GRDC_2589752', 'GRDC_2589753', 'GRDC_2589800', 'GRDC_2590220', 'GRDC_2590300', 'GRDC_2911206', 'GRDC_2969014', 'GRDC_3268100', 'GRDC_3275280', 'GRDC_3469400', 'GRDC_3650470', 'GRDC_4119201', 'GRDC_4119317', 'GRDC_4121300', 'GRDC_4121301', 'GRDC_4122350', 'GRDC_4122636', 'GRDC_4122900', 'GRDC_4126805', 'GRDC_4133501', 'GRDC_4150451', 'GRDC_4150500', 'GRDC_4151801', 'GRDC_4152601', 'GRDC_4208515', 'GRDC_4208516', 'GRDC_4236700', 'GRDC_4244460', 'GRDC_4244770', 'GRDC_4355500', 'GRDC_4356050', 'GRDC_4356800', 'GRDC_4358800', 'GRDC_4874750', 'GRDC_5101130', 'GRDC_5202070', 'GRDC_5204104', 'GRDC_5654140', 'GRDC_5868200', 'GRDC_5868300', 'GRDC_6111100', 'GRDC_6142150', 'GRDC_6233690', 'GRDC_6233710', 'GRDC_6335240', 'GRDC_6335301', 'GRDC_6335303', 'GRDC_6340190', 'GRDC_6342510', 'GRDC_6342925', 'GRDC_6421102', 'GRDC_6444090', 'GRDC_6563200', 'GRDC_6607651', 'GRDC_6688650', 'GRDC_6691650', 'GRDC_6731260', 'GRDC_6742600', 'GRDC_6935020', 'GRDC_6935145', 'GRDC_6935310', 'GRDC_6942100', 'GRDC_6948100', 'GRDC_6972200', 'GRDC_6976450', 'USGS_01011000', 'USGS_01017000', 'USGS_01029500', 'USGS_01030500', 'USGS_01034000', 'USGS_01054000', 'USGS_01054500', 'USGS_01059000', 'USGS_01076500', 'USGS_01092000', 'USGS_01200500', 'USGS_01321000', 'USGS_01325000', 'USGS_01421000', 'USGS_01425000', 'USGS_01426500', 'USGS_01427510', 'USGS_01428500', 'USGS_01454700', 'USGS_01562000', 'USGS_01563500', 'USGS_01668000', 'USGS_02019500', 'USGS_02024752', 'USGS_02025500', 'USGS_02041650', 'USGS_02075500', 'USGS_02083500', 'USGS_02084000', 'USGS_02089500', 'USGS_02098198', 'USGS_02105769', 'USGS_02151500', 'USGS_02163001', 'USGS_02181580', 'USGS_02192000', 'USGS_02210500', 'USGS_02215000', 'USGS_02223000', 'USGS_02244440', 'USGS_02246025', 'USGS_02312720', 'USGS_02312762', 'USGS_02319800', 'USGS_02320000', 'USGS_02320500', 'USGS_02323000', 'USGS_02323500', 'USGS_02334430', 'USGS_02336000', 'USGS_02336490', 'USGS_02337170', 'USGS_02346180', 'USGS_02347500', 'USGS_02349605', 'USGS_02352500', 'USGS_02359051', 'USGS_02365500', 'USGS_02372422', 'USGS_02404400', 'USGS_02414500', 'USGS_02414715', 'USGS_02418500', 'USGS_02474560', 'USGS_02486000', 'USGS_02489000', 'USGS_03011020', 'USGS_03056000', 'USGS_03146500', 'USGS_03147500', 'USGS_03165500', 'USGS_03192000', 'USGS_03221000', 'USGS_03263000', 'USGS_03324500', 'USGS_03351000', 'USGS_03453500', 'USGS_03455000', 'USGS_03503000', 'USGS_03604400', 'USGS_04040000', 'USGS_04059000', 'USGS_04065722', 'USGS_04066800', 'USGS_04067500', 'USGS_04069416', 'USGS_04114000', 'USGS_04116000', 'USGS_04137005', 'USGS_05369000', 'USGS_05395000', 'USGS_05398000', 'USGS_05427085', 'USGS_05453520', 'USGS_05454500', 'USGS_05481650', 'USGS_05520500', 'USGS_05545750', 'USGS_05599500', 'USGS_06037500', 'USGS_06038800', 'USGS_06099500', 'USGS_06191500', 'USGS_06192500', 'USGS_06335500', 'USGS_06336000', 'USGS_06337000', 'USGS_06620000', 'USGS_06630000', 'USGS_06646800', 'USGS_06656000', 'USGS_06793000', 'USGS_06810000', 'USGS_06817000', 'USGS_06817500', 'USGS_06817700', 'USGS_07052500', 'USGS_07061000', 'USGS_07076517', 'USGS_07094500', 'USGS_07099400', 'USGS_07099970', 'USGS_07119700', 'USGS_07124000', 'USGS_07151000', 'USGS_07158000', 'USGS_07159100', 'USGS_07170990', 'USGS_07171000', 'USGS_07176000', 'USGS_07188000', 'USGS_07196500', 'USGS_07198000', 'USGS_07328100', 'USGS_07331000', 'USGS_07340000', 'USGS_07348000', 'USGS_08065000', 'USGS_08065350', 'USGS_08319000', 'USGS_08331160', 'USGS_08331510', 'USGS_08384500', 'USGS_09093700', 'USGS_09106150', 'USGS_09205000', 'USGS_09209400', 'USGS_09217000', 'USGS_09346400', 'USGS_09365000', 'USGS_09368000', 'USGS_09512406', 'USGS_12027500', 'USGS_12414500', 'USGS_12420500', 'USGS_12422500', 'USGS_12510500', 'USGS_13011000', 'USGS_13018750', 'USGS_13056500', 'USGS_13183000', 'USGS_13247500', 'USGS_13249500', 'USGS_13340600', 'USGS_14148000', 'USGS_14150000', 'USGS_14166000', 'USGS_14231000', 'USGS_14238000']

# === Step 0: è¯»å–æ•°æ® ===
df_all = pd.read_csv('node_clip_0403.csv')
df_comid = pd.read_csv('../2-preprocess/3.gauge3000_q50_slp.csv')[['stationid','COMID']]
df_all = df_all.replace(-999999999999, np.nan).dropna()
df_all = df_all.merge(df_comid, on='stationid', how='inner')

# å»é™¤èŠ‚ç‚¹æ•°æ®å°‘äº10çš„
df_all['node_data_num'] = df_all.groupby('node_id')['lat'].transform('count')
df_all = df_all[df_all['node_data_num'] >= 10]
stationids = df_all['stationid'].unique()
# é€‰æ¯ä¸ª station çš„æœ€å¤§ rank_corr èŠ‚ç‚¹
df_node = df_all[['node_id', 'stationid']].drop_duplicates(subset='node_id')

df_node = df_node.set_index('node_id')
for node in df_node.index:
    df_sel = df_all[df_all['node_id']==node]
    df_node.loc[node,'rank_corr'] = spearmanr(df_sel['width'], df_sel['wse'])[0]
df_node = df_node.reset_index()

max_idx = df_node.groupby('stationid')['rank_corr'].idxmax()
df_node_rmax = df_node.loc[max_idx]
df_all = df_all[df_all['node_id'].isin(df_node_rmax['node_id'])]

df_w = pd.read_csv('1.width_range.csv', index_col='stationid')
df_final = []


for s in stationids[:]: 
    print(s)
    df = df_all[df_all['stationid'] == s].copy()

    if s not in df_w.index: 
        continue

    # Step 1: ä¸ç¡®å®šåº¦é˜ˆå€¼ç­›é€‰
    df['width_u_r'] = df['width_u'] / df['width']
    df1 = df[(df['wse_u'] <= 0.4) & (df['width_u_r'] <= 0.1)]

    if len(df1) < 5:
        if s in target_stations:
            print(f'{s} step1')
        continue
    
    # Step 2: é¡ºåºä¸€è‡´æ€§å‰”é™¤
    while True:
        df2 = df1.copy()
        n = len(df2)
        w = df2['width'].values
        h = df2['wse'].values
        w_diff = np.repeat(w, n).reshape(n, -1) - np.tile(w, (n, 1))
        h_diff = np.repeat(h, n).reshape(n, -1) - np.tile(h, (n, 1))
        wh = w_diff * h_diff
        inverse = np.count_nonzero(wh < 0, axis=1)
        idx_max = np.argmax(inverse)
        if inverse[idx_max] / n < 0.5: 
            if s in target_stations:
                print(f'{s} step2')
            break
        df2 = df2.drop(df2.index[idx_max])


    # Step 3: ç¦»ç¾¤å€¼å‰”é™¤
    w_low, w_high = df_w.loc[s, ['w_low', 'w_high']]
    d_bankfull = 0.27 * (w_high / 7.2)**0.6
    h50 = df2['wse'].median()
    df3 = df2[(df2['wse'] <= h50 +  d_bankfull) & (df2['wse'] >= h50 - d_bankfull)]
    if len(df3) < 5:
        if s in target_stations:
            print(f'{s} step3')
        continue 
        
    df_final.append(df3)

    if not DRAW_FIGURE:
        continue

    
    # æ‰¾å‡ºæ¯æ­¥å‰”é™¤çš„ç‚¹
    s1_drop = df[~df.index.isin(df1.index)]
    s2_drop = df1[~df1.index.isin(df2.index)]
    s3_drop = df2[~df2.index.isin(df3.index)]

    # å¯è§†åŒ–ä¸‰æ­¥ç»“æœ
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.errorbar(df['width'], df['wse'], xerr=df['width_u'], yerr=df['wse_u'], fmt='none', lw=0.5, ecolor='lightgray', zorder=0)

    ax.scatter(s1_drop['width'], s1_drop['wse'], c='blue', s=12, label='Step 1 dropped', alpha=0.6, marker='x')
    ax.scatter(s2_drop['width'], s2_drop['wse'], c='green', s=12, label='Step 2 dropped', alpha=0.6, marker='^')
    ax.scatter(s3_drop['width'], s3_drop['wse'], c='orange', s=12, label='Step 3 dropped', alpha=0.6, marker='s')
    ax.scatter(df3['width'], df3['wse'], c='red', s=12, label='Retained', alpha=0.8, marker='o')

    ax.set_xlabel('Width (m)')
    ax.set_ylabel('WSE (m)')
    ax.legend()
    ax.set_title(f'{s} - Data QC steps')
    fig.tight_layout()
    plt.savefig(f'{folder}/{s}.png', dpi=200)
    plt.close()

# === è¾“å‡ºç»“æœ ===
df_final = pd.concat(df_final)
df_final = df_final.drop_duplicates(subset=['node_id', 'time', 'stationid'])  # æˆ–ä»…ä½¿ç”¨ .drop_duplicates()
df_final.reset_index(drop=True, inplace=True)
df_final.insert(0, 'COMID', df_final.pop('COMID'))
df_final.to_csv('2.swot-points-selection.csv', index=False)
```
