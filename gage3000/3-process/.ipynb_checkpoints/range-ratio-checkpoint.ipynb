```python
import pandas as pd

# ========== 数据集1: VersionD ==========
swot_df_1 = pd.read_csv('1.all_matched_points_VersionD.csv')
range_df_1 = pd.read_csv('half-processed/3.fit_proba_modified_q50_bins_node_0.03_noqa_VersionD.csv')
stationids = pd.read_csv('ori/5.q_kge_med_modified_q50.csv')['stationid']

swot_df_1 = swot_df_1[swot_df_1['stationid'].isin(stationids)][['stationid', 'width']]
range_df_1 = range_df_1[range_df_1['stationid'].isin(stationids)][['stationid', 'w_high', 'w_low']]

# ========== 数据集2: ori ==========
swot_df_2 = pd.read_csv('ori/node_clip_0403.csv')
range_df_2 = pd.read_csv('ori/3.fit_proba_modified_q50.csv')

swot_df_2 = swot_df_2[swot_df_2['stationid'].isin(stationids)][['stationid', 'width']]
range_df_2 = range_df_2[range_df_2['stationid'].isin(stationids)][['stationid', 'w_high', 'w_low']]

# ========== 计算函数 ==========
def calc_range_stats(swot_df, range_df, stationids, suffix):
    results = []
    for sid in stationids:
        swot_vals = swot_df.loc[swot_df['stationid'] == sid, 'width']
        range_vals = range_df.loc[range_df['stationid'] == sid]
        
        if len(swot_vals) > 0 and len(range_vals) > 0:
            w_low = range_vals['w_low'].iloc[0]
            w_high = range_vals['w_high'].iloc[0]
            
            if w_high > w_low:
                in_range = swot_vals.between(w_low, w_high)
                if in_range.any():
                    in_range_vals = swot_vals[in_range]
                    actual_range = in_range_vals.max() - in_range_vals.min()
                    range_width = w_high - w_low
                    
                    results.append({
                        'stationid': sid,
                        f'w_high_{suffix}': w_high,
                        f'w_low_{suffix}': w_low,
                        f'actual_range_{suffix}': actual_range,
                        f'range_width_{suffix}': range_width,
                        f'range_ratio_{suffix}': actual_range / range_width
                    })
    
    return pd.DataFrame(results).drop_duplicates(subset='stationid')

# ========== 计算两个数据集的结果 ==========
result_1 = calc_range_stats(swot_df_1, range_df_1, stationids, '1')
result_2 = calc_range_stats(swot_df_2, range_df_2, stationids, '2')

# ========== 合并结果 ==========
merged_df = pd.merge(result_1, result_2, on='stationid', how='outer')

# 重新排列列顺序
cols = ['stationid', 
        'w_high_1', 'w_low_1', 'actual_range_1', 'range_width_1', 'range_ratio_1',
        'w_high_2', 'w_low_2', 'actual_range_2', 'range_width_2', 'range_ratio_2']
merged_df = merged_df[cols]

# ========== 输出结果 ==========
print("=" * 80)
print("数据集1 (VersionD) 统计:")
print(f"  站点数量: {len(result_1)}")
print(f"  范围比例中位数: {result_1['range_ratio_1'].median():.3f}")
print(f"  范围比例平均值: {result_1['range_ratio_1'].mean():.3f}")

print("\n数据集2 (ori) 统计:")
print(f"  站点数量: {len(result_2)}")
print(f"  范围比例中位数: {result_2['range_ratio_2'].median():.3f}")
print(f"  范围比例平均值: {result_2['range_ratio_2'].mean():.3f}")

print("\n" + "=" * 80)
print(f"合并表格 (共 {len(merged_df)} 个站点):")
print(merged_df.to_string(index=False))

# 保存到CSV
merged_df.to_csv('range_ratio_comparison.csv', index=False)
print("\n结果已保存到: range_ratio_comparison.csv")
```
