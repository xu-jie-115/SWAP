```python
import geopandas as gpd
import pandas as pd
import glob,os

att_csv = pd.read_csv('GAGE_3000pointsbuffer_GLOW_points_matched.csv')[['ID','stationid']]
csv_files = glob.glob('../GLOW_Width/GLOW_width_region_*.csv')  # You can change the directory if needed

# List to hold the merged DataFrames
merged_dfs = []

# Loop through all CSV files
for file_path in csv_files:
    # Read the current CSV file
    width = pd.read_csv(file_path)
    
    # Merge the current CSV file with att_csv on the 'ID' column (right join)
    width_timeseries = att_csv.merge(width, on='ID', how='left')
    
    # Append the result to the merged_dfs list
    merged_dfs.append(width_timeseries)

# Concatenate all merged DataFrames into one DataFrame
final_df = pd.concat(merged_dfs, ignore_index=True)
```

```python
final_df = final_df.dropna(subset=['date'])
# Save the final merged DataFrame to a CSV file
final_df.to_csv("gages3000_glow_width_timeseries.csv", index=False)
average_width_df = final_df.groupby(['stationid','date'])['width'].mean().reset_index()
average_width_df.to_csv("gages3000_glow_datemean_width_timeseries.csv", index=False)
```
