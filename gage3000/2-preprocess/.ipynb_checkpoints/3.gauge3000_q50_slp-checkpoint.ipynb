```python
import glob
import numpy as np
import pandas as pd
import geopandas as gpd

# q50_df = pd.read_csv('gages3000_GRFR_q50.csv')
# 获取所有匹配的shp文件
shp_files = glob.glob('/shared1/RESEARCH_DATA/MERIT_Basins/riv_pfaf_*.shp')

# 读取所有shp文件并合并
gdfs = [gpd.read_file(f) for f in shp_files]
dfll = gpd.GeoDataFrame(pd.concat(gdfs, ignore_index=True))
# 丢掉 geometry 列
dfll_no_geometry = dfll.drop(columns='geometry')

# 保存为 CSV 文件
dfll_no_geometry.to_csv('/home/xj/device/data/Merit_riv_att.csv', index=False)
```

```python
q50_df = pd.read_csv('2.gages3000_GRFR_q50.csv')
q50_slp_df = dfll_no_geometry[['COMID','slope']].merge(q50_df,on='COMID',how='right')
q50_slp_df.to_csv('3.gages3000_GRFR_q50_slp.csv')
```
