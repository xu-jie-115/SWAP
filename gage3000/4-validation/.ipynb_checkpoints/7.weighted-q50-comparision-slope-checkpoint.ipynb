```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as patches
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr
import numpy as np


plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 16
plt.rcParams['axes.titlesize'] = 24
plt.rcParams['axes.labelsize'] = 24
plt.rcParams['axes.titlepad'] = 10
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['xtick.major.width'] = 1.6
plt.rcParams['ytick.major.width'] = 1.6

slp_df = pd.read_csv('../2-preprocess/3.gauge3000_q50_slp.csv')[['stationid','slope']]

# 计算slope列的10%分位数
slope_10th_percentile = slp_df['slope'].quantile(q=0.1)
print("slope的10%分位数:", slope_10th_percentile)

swap_df = pd.read_csv('2.validation_SWAP_modified_q50.csv')
swap_df.rename(columns={'PB': 'pBIAS'}, inplace=True)
swap_df = swap_df.merge(slp_df,on ='stationid',how = 'inner')
swap_df = swap_df.drop_duplicates()
swap_df = swap_df[swap_df['slope']<slope_10th_percentile]

w83_df = pd.read_csv('1.validation_W83_q50.csv')
w83_df.rename(columns={'PB': 'pBIAS'}, inplace=True)
w83_df = w83_df[w83_df['stationid'].isin(swap_df['stationid'])]


columns = ['KGE', 'NSE', 'NRMSE', 'CC', 'pBIAS', 'RV']
# 删除包含缺失值的行
swap_df.dropna(subset=columns, inplace=True)
w83_df.dropna(subset=columns, inplace=True)


# Define x-axis ranges for each subplot
x_axis_ranges = {
    'KGE': [-1, 1],
    'NSE': [-1, 1],
    'NRMSE': [0, 1],
    'CC': [0, 1],
    'pBIAS': [-1, 1],
    'RV': [0, 2]
}

# Set up the figure and axes with different colors for each label
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# Define a function to plot CDF with different colors
def plot_cdf(data, ax, label, x_range=None, color=None):
    sorted_data = sorted(data)
    cdf = [i / len(sorted_data) for i in range(1, len(sorted_data) + 1)]
    ax.plot(sorted_data, cdf, label=label, color=color)

    # Set x-axis limits
    if x_range:
        ax.set_xlim(x_range)
    
    # ax.set_xlabel(label)
    ax.set_ylabel('CDF')

# Define distinct colors for each dataset
colors = {
    'BAM': 'blue',
    'Geobam': 'green',
    'W83': 'red',
    'SWAP': 'orange'
}

# Plot CDF for each column with the specified colors
for i, col in enumerate(columns):
    ax = axes[i//3, i%3]  # Calculate the subplot position
    plot_cdf(w83_df[col], ax, 'W83', x_range=x_axis_ranges[col], color=colors['W83'])
    plot_cdf(swap_df[col], ax, 'SWAP', x_range=x_axis_ranges[col], color=colors['SWAP'])
    ax.legend()
    ax.set_title(col)

# Adjust layout
plt.tight_layout()
plt.savefig('7.comparing-CDF-modified_q50-slope.png',dpi = 500)
plt.show()
```

```python
def compute_stats(df, name):
    total = len(df)
    results = {
        'method': name,
        'KGE>0': (df['KGE'] > 0).sum() / total,
        'NSE>0': (df['NSE'] > 0).sum() / total,
        '|pBIAS|<0.5': (df['pBIAS'].abs() < 0.5).sum() / total,
        '|RV-1|<0.5': ((df['RV'] - 1).abs() < 0.5).sum() / total,
        'NRMSE<0.6': (df['NRMSE'] < 0.6).sum() / total,
        'CC>0.5': (df['CC'] > 0.5).sum() / total,
        'NRMSE_median': df['NRMSE'].median(),
        'NRMSE_mean': df['NRMSE'].mean(),
    }
    return results

# 对每个模型分别统计
results = []
results.append(compute_stats(swap_df, 'SWAP'))
results.append(compute_stats(w83_df, 'W83'))

# 转为 DataFrame 展示
import pandas as pd
summary_df = pd.DataFrame(results)
pd.set_option('display.float_format', '{:.2%}'.format)  # 百分比格式显示
print(summary_df)
```
