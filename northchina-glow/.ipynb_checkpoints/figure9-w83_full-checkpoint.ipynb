```python
import os,glob
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr
import matplotlib as mpl
import matplotlib.ticker as ticker
```

```python
df_q50 = pd.read_csv('./2-preprocess/4.q50_weighted_slope.csv')
folder_path ='./4-validation/observation'

def KGE(y_true, y_pred):  # improved 2012
    correlation = np.corrcoef(y_true, y_pred)[0, 1]
    alpha = (np.std(y_pred)/np.mean(y_pred)) / (np.std(y_true)/np.mean(y_true))
    beta = np.mean(y_pred) / np.mean(y_true)
    return 1 - np.sqrt((correlation - 1)**2 + (alpha - 1)**2 + (beta - 1)**2)


# 定义计算NSE的函数
def NSE(observed, simulated):
    return 1 - np.sum((observed - simulated) ** 2) / np.sum((observed - np.mean(observed)) ** 2)


# 定义计算rRMSE的函数
def nRMSE(observed, simulated):
    rmse = np.sqrt(mean_squared_error(observed, simulated))
    return rmse / np.mean(observed)

# 定义计算CC的函数
def CC(observed, simulated):
    return pearsonr(observed, simulated)[0]  # 使用Spearman相关系数作为CC

# 创建一个结果存储的DataFrame
results = []

df_q83 = pd.DataFrame()

stations = df_q50['stationid'].unique()
for station in stations:
    print(station)
    file_match = glob.glob(os.path.join(folder_path, f"*{station}*.csv"))
    if not file_match:
        print(f"File for station {station} not found. Skipping...")
        continue
    file_path = file_match[0]
    df_L8 = pd.read_csv(file_path).dropna(subset=['Q'])[['date','correct_maxRiverWidth','Q']]
    df_L8 = df_L8.rename(columns= {'correct_maxRiverWidth':'width'})
    q50 = df_q50.loc[df_q50['stationid'] == station, 'q50_weighted'].values[0]
    w_mean = df_L8['width'].median()
    a      = q50/(w_mean**(8.0/3.0))
    df_L8['83Q'] = a*df_L8['width']**(8.0/3.0)
    df_L8['stationid']= station   
    df_q83 = pd.concat([df_q83,df_L8], ignore_index=True)
    
    observed = df_L8['Q']
    simulated = df_L8['83Q']
    
    # 计算各个指标
    cc = CC(observed, simulated)
    nrmse = nRMSE(observed, simulated)
    kge = KGE(observed, simulated)
    nse = NSE(observed, simulated)
    pbias = (simulated.mean() / observed.mean() - 1) 
    rv    = simulated.std() / simulated.mean()/ (observed.std() / observed.mean())
    results.append({'stationid': station, 'KGE': kge, 'NSE': nse,'nRMSE': nrmse,'CC': cc, 'pBIAS': pbias, 'RV': rv})


results_df = pd.DataFrame(results)
results_df.to_csv('./validation_83.csv')
df_q83.to_csv('./q_83.csv')
print('df_q83',df_q83)
```

```python
swap_df = pd.read_csv('./3-process/5.q_kge_med_modified_q50.csv')
df_q83_datematch = df_q83.merge(swap_df[['stationid', 'date']], on=['stationid', 'date'], how='inner').drop_duplicates(subset = ['stationid', 'date'])

df_q83_datematch = df_q83_datematch.drop_duplicates(subset = ['stationid','date'])

results = []

stations = df_q83_datematch['stationid'].unique()
for station in stations:
    observed  = df_q83_datematch[df_q83_datematch['stationid']==station]['Q']
    simulated = df_q83_datematch[df_q83_datematch['stationid']==station]['83Q']
    
    # 计算各个指标
    cc = CC(observed, simulated)
    nrmse = nRMSE(observed, simulated)
    kge = KGE(observed, simulated)
    nse = NSE(observed, simulated)
    pbias = (simulated.mean() / observed.mean() - 1) 
    rv    = simulated.std() / simulated.mean()/ (observed.std() / observed.mean())
    results.append({'stationid': station, 'KGE': kge, 'NSE': nse,'nRMSE': nrmse,'CC': cc, 'pBIAS': pbias, 'RV': rv})

w83_df = pd.DataFrame(results)
w83_df = w83_df.sort_values('stationid')
w83_df.to_csv('./validation_83_datematch.csv')
```

```python
w83_df  = w83_df[w83_df['stationid'].isin(swap_df['stationid'])].reset_index(drop=True)
swap_df = swap_df.drop_duplicates(subset = ['stationid'])
swap_df = swap_df.sort_values('stationid').reset_index(drop=True)
w83_df  = w83_df[~w83_df['stationid'].isin(['longmenzhen'])]
swap_df = swap_df[~swap_df['stationid'].isin(['longmenzhen'])]

print('w83_df',w83_df[~w83_df['stationid'].isin(['longmenzhen'])])
print('swap_df',swap_df[~swap_df['stationid'].isin(['longmenzhen'])])
def compute_stats(df, name):
    total = len(df)
    results = {
        'method': name,
        'KGE>0': (df['KGE'] > 0).sum() / total,
        'NSE>0': (df['NSE'] > 0).sum() / total,
        '|pBIAS|<0.5': (df['pBIAS'].abs() < 0.5).sum() / total,
        '|RV-1|<0.5': ((df['RV'] - 1).abs() < 0.5).sum() / total,
        'NRMSE<0.6': (df['nRMSE'] < 0.6).sum() / total,
        'CC>0.5': (df['CC'] > 0.5).sum() / total,
        'NRMSE_median': df['nRMSE'].median(),
        'NRMSE_mean': df['nRMSE'].mean(),
    }
    return results

# 对每个模型分别统计
results = []
results.append(compute_stats(swap_df, 'SWAP'))
results.append(compute_stats(w83_df, 'W83'))

# 转为 DataFrame 展示
import pandas as pd
summary_df = pd.DataFrame(results)
pd.set_option('display.float_format', '{:.2%}'.format)  # 百分比格式显示
print(summary_df)




com_df  = swap_df[['KGE', 'NSE','nRMSE','CC', 'pBIAS', 'RV']] - w83_df[['KGE', 'NSE','nRMSE','CC', 'pBIAS', 'RV']]
com_df['|pBIAS|'] = abs(swap_df[['pBIAS']])- abs(w83_df[['pBIAS']])
com_df['|RV-1|'] = abs(swap_df[['RV']]-1)- abs(w83_df[['RV']]-1)
com_df['stationid'] = swap_df['stationid']
com_df = com_df[~com_df['stationid'].isin(['longmenzhen'])].reset_index(drop=True)
com_df['stationid'] = com_df['stationid'].str.capitalize()
# com_df ['stationid'] = com_df ['stationid'].replace({'changjiangtun':'Changjiangtun',
#                                                'dalai':'Dalai',
#                                                'fuyu':'Fuyu',
#                                                'heishiguan':'Heishiguan',
#                                                'huaibin':'Huaibin',
#                                                'humaqiao':'Humaqiao',
#                                                'jiamusi':'Jiamusi',
#                                                'jiangjiaji':'Jiangjiaji',
#                                                'lianhua':'Lianhua',
#                                                'liujiatun':'Liujiatun',
#                                                'luanxian':'Luanxian',
#                                                'shihuiyao':'Shihuiyao',
#                                                'shuanghetun':'Shuanghetun',
#                                                'tangmazhai':'Tangmazhai',
#                                                'tieling':'Tieling',
#                                                'tonghe':'Tonghe',
#                                                'tonghua':'Tonghua',
#                                                'xingjiawopeng':'Xingjiawopeng'})
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as patches


plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 16
plt.rcParams['axes.titlesize'] = 24
plt.rcParams['axes.labelsize'] = 24
plt.rcParams['axes.titlepad'] = 10
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['xtick.major.width'] = 1.6
plt.rcParams['ytick.major.width'] = 1.6


heatmap_data = com_df[['stationid', 'KGE', 'NSE','CC','nRMSE', '|pBIAS|', '|RV-1|']]

# 设置 'short-name' 为索引
heatmap_data.set_index('stationid', inplace=True)
heatmap_data.index = heatmap_data.index.str.capitalize()

filtered_df = com_df[(com_df['KGE'] > 0) & (com_df['NSE'] > 0) &(com_df['nRMSE'] < 0)]

filtered_com = com_df[(com_df['KGE'] < 0) & (com_df['NSE'] < 0) &(com_df['nRMSE'] > 0)]


# 设置绘图大小
plt.figure(figsize=(10, 6))

# 绘制热图，设置 vmin 和 vmax，控制 colorbar 范围
ax = sns.heatmap(heatmap_data, cmap='coolwarm', annot=False, center=0, 
                 vmin=-1, vmax=1, cbar_kws={'label': 'Value', 'extend': 'both'})

# Get the station names that meet the condition
stations_with_circle = filtered_df['stationid'].values
# Add circles on the stations that satisfy the condition
for station in stations_with_circle:
    # print(f"Looking for station: {station}")
    # print(f"Available stations in heatmap_data index: {heatmap_data.index}")
    # Get the row index of the station in the heatmap (now it's the index of the DataFrame)
    row_idx = np.where(heatmap_data.index == station)[0][0]
    # In the heatmap, rows are the features, so find the column number for a feature like 'KGE'
    col_idx = list(heatmap_data.columns).index('KGE')  # Change 'KGE' to whichever feature you want to highlight
    # Add a circle marker at the corresponding location in the heatmap
    ax.scatter(0.1, row_idx + 0.5, s=100, facecolors='none', edgecolors='black', linewidths=2)


# Get the station names that meet the condition
stations_with_circle = filtered_com['stationid'].values
# Add circles on the stations that satisfy the condition
for station in stations_with_circle:
    # Get the row index of the station in the heatmap (now it's the index of the DataFrame)
    row_idx = np.where(heatmap_data.index == station)[0][0]
    # In the heatmap, rows are the features, so find the column number for a feature like 'KGE'
    col_idx = list(heatmap_data.columns).index('KGE')  # Change 'KGE' to whichever feature you want to highlight
    # Add a circle marker at the corresponding location in the heatmap
    ax.scatter(0.1, row_idx + 0.5, s=100, facecolors='black', edgecolors='black', linewidths=2)

# 设置标题
plt.title('Heatmap of CC, nRMSE, KGE, NSE across short-names')

# 显示图形
plt.tight_layout()
plt.savefig('./Fig9.png',dpi = 500)
plt.show()
```
