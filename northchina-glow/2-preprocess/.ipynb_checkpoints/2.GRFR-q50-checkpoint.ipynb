```python
import glob
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
from netCDF4 import Dataset
import geopandas as gpd

# 获取所有匹配的 nc 文件
final_df = pd.read_csv('2.GRFR_q50_04.csv')
# 合并所有 DataFrame
north_20 = gpd.read_file('1.north_GLOW_buffer_matched_points.shp')
df_merged = north_20.merge(final_df,on='COMID',how='left')
df_merged = df_merged.rename(columns={'stnmpy': 'stationid'})
# 保存合并后的数据
df_merged.to_csv('2.north20_GRFR_q50.csv', index=False)
print("合并完成，已保存为 merged_q50_all.csv")
print(df_merged.columns)
```
