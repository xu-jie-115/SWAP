```python
import geopandas as gpd
import pandas as pd

glow_width = pd.read_csv('/home/xj/device/data/GLOW/GLOW_Width/GLOW_width_all.csv')
print(glow_width.columns)
polygon_match = gpd.read_file('/home/xj/device/data/GLOW/GLOW_Cross_section/all.shp')
print(polygon_match.columns)
```

```python
glow_points = polygon_match 
matched_polygons = gpd.read_file('../1-buffer-match/GAGE_3000points_MERIT_Basin_buffer_polygon_4widths.shp')

matched_polygons = matched_polygons.to_crs(glow_points.crs)

# 进行空间连接，保留在 polygon 内的点
points_in_polygons = gpd.sjoin(glow_points, matched_polygons, how="inner", predicate="within").drop(columns=['index','index_right'])

points_in_polygons.rename(columns={
    'lat_left': 'glow_lat',
    'lon_left': 'glow_lon',
    'lat_right':'gage_lat',
    'lon_right':'gage_lon',
    # 'distance':'GRGEpoints-distance-COMID',
    'width':'COMID-meanwidth',
}, inplace=True)

# 保存结果到新的 Shapefile
points_in_polygons.to_file("1.GAGE_3000pointsbuffer_GLOW_points_matched_points.shp")
print(points_in_polygons.columns)
```
