```python
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score, mean_squared_error


df_swot_all = pd.read_csv('../3-process/2.swot-points-selection_iqr_node_1.5_noqa_VersionD.csv')
df_fit_all = pd.read_csv('../3-process/3.fit_proba_modified_q50_iqr_node_1.5_noqa_VersionD.csv')
df_med_all = pd.read_csv('../3-process/4.hypso_med_modified_q50_iqr_node_1.5_noqa_VersionD.csv')
stations = sorted(df_fit_all['stationid'].unique())

df_res = pd.DataFrame({'stationid':stations}).set_index('stationid')
for s in stations:
    df_swot = df_swot_all[df_swot_all['stationid']==s]
    df_fit = df_fit_all[df_fit_all['stationid']==s]
    df_med = df_med_all[df_med_all['stationid']==s]
    
    w_low, w_high, slp = df_fit.iloc[0][['w_low','w_high','slp']]
    d_bankfull = 0.27 * (w_high / 7.2)**0.6
    
    df_swot['wse_hypso'] = np.interp(df_swot['width'], df_med['width'], df_med['wse'])
    df_res.loc[s,'r2'] = r2_score(df_swot['wse'], df_swot['wse_hypso'])
    df_res.loc[s,'rmse'] = mean_squared_error(df_swot['wse'], df_swot['wse_hypso'])**0.5
    df_res.loc[s,'nrmse'] = df_res.loc[s,'rmse'] / d_bankfull

breakpoint()
df_res.to_csv('swot_hypso_metrics.csv')
```
